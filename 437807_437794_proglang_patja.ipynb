```python
def hasNext():
    global i
    global siz
    if(i >= size):
        return False
    else: 
        return True
        
def has():
    global i
    global size
    if(i >= size):
        return False
    else: 
        try:
            return s[i+1]
        except:
            return '-'
        
def getNext():
    global i
    return s[i+1]

def getC(s):
    global i
    return s[i]

def token(s):
    global i
    global size
    global nl
    global goto
    global nextGoto
    global nextNl
    state = 0
    size = len(s)
    value = ''
    
    for j in range (size+1):
        i = i+1
        #print('round ' + str(j), 'state ' + str(state), 'char '+ str(i))
        if(state == 0):
            if(getC(s) == 'p' or getC(s) == 'P'):
                state = 1
                continue
            if(getC(s) == 'i' or getC(s) == 'I'):
                state = 6
                continue
            if(getC(s) == 'g' or getC(s) == 'G'):
                state = 8
                continue
            if(getC(s) == 's' or getC(s) == 'S'):
                state = 12
                continue
            if(getC(s) == '+'):
                state = 16
                continue
            if(getC(s) == '-'):
                state = 17
                continue
            if(getC(s) == '='):
                state = 18
                continue
            if(getC(s) == '<'):
                state = 19
                continue
            if(getC(s) in string.digits):
                value += getC(s)
                state = 20
                continue
            if(getC(s) in string.ascii_uppercase):
                state = 21
                continue
        elif(state == 1):
            if(not hasNext()):
                return ('#id', ord(s[i-1])-64) #id
            if(getC(s) == 'r' or getC(s) == 'R'):
                state = 2
                continue
            else:
                return 0
        elif(state == 2):
            if(not hasNext()):
                return 0
            if(getC(s) == 'i' or getC(s) == 'I'):
                state = 3
                continue
            else:
                return 0
        elif(state == 3):
            if(not hasNext()):
                return 0
            if(getC(s) == 'n' or getC(s) == 'N'):
                state = 4
                continue
            else:
                return 0
        elif(state == 4):
            if(not hasNext()):
                return 0
            if(getC(s) == 't' or getC(s) == 'T'):
                state = 5
                continue
            else:
                return 0
        elif(state == 5):
            if(not hasNext()):
                return ('#print', 0) #print
            else:
                return 0
        elif(state == 6):
            if(not hasNext()):
                return ('#id', ord(s[i-1])-64) #id
            if(getC(s) == 'f' or getC(s) == 'F'):
                state = 7
                continue
            else:
                return 0
        elif(state == 7):
            if(not hasNext()):
                nextGoto = True
                return ('#if', 0) #if
            else:
                return 0
        elif(state == 8):
            if(not hasNext()):
                return ('#id', ord(s[i-1])-64) #id
            if(getC(s) == 'o' or getC(s) == 'O'):
                state = 9
                continue
            else:
                return 0
        elif(state == 9):
            if(not hasNext()): 
                return 0
            if(getC(s) == 't' or getC(s) == 'T'):
                state = 10
                continue
            else:
                return 0
        elif(state == 10):
            if(not hasNext()):
                return 0
            if(getC(s) == 'o' or getC(s) == 'O'):
                state = 11
                continue
            else:
                return 0
        elif(state == 11):
            if(not hasNext()):
                return ('#goto', 0) #goto
            else:
                return 0
        elif(state == 12):
            if(not hasNext()):
                return ('#id', ord(s[i-1])-64) #id
            if(getC(s) == 't'  or getC(s) == 'T'): state = 13
            else:
                return 0
        elif(state == 13):
            if(not hasNext()):
                return 0
            if(getC(s) == 'o' or getC(s) == 'O'): state = 14
            else:
                return 0
        elif(state == 14):
            if(not hasNext()):
                return 0
            if(getC(s) == 'p' or getC(s) == 'P'): state = 15
            else:
                return 0
        elif(state == 15):
            if(not hasNext()):
                return ('#stop', 0) #stop
            else:
                return 0
        elif(state == 16):
            if((not hasNext()) or getNext() in string.ascii_uppercase or getNext() in string.digits):
                return ('#op', 1) #+
            else:
                return 0
        elif(state == 17):
            if((not hasNext()) or getNext() in string.ascii_uppercase or getNext() in string.digits):
                return ('#op', 2) #-
            else:
                return 0
        elif(state == 18):
            if((not hasNext()) or getNext() in string.ascii_uppercase or getNext() in string.digits):
                return ('#op', 4) #=
            else:
                return 0
        elif(state == 19):
            if((not hasNext()) or getNext() in string.ascii_uppercase or getNext() in string.digits):
                return ('#op', 3) #<
            else:
                return 0
        elif(state == 20):
            if(not hasNext()):
                if(int(value) > 1000):
                    value = 0
                    return 0
                elif(nl):
                    nl = False
                    return ('#lin_num', int(value)) #line
                elif(nextGoto and nextNl):
                    nextGoto = False
                    return ('#lin_num', int(value)) #goto
                else:
                    return ('#const', int(value)) #const
            else:
                dig = getC(s)
                if(dig in string.digits):
                    value += dig
                    state = 20
                    continue
                else:
                    return 0
        elif(state == 21):
            if(not hasNext()): 
                return ('#id',ord(s[i-1])-64) #id
            else:
                return 0
```

```python
grammar = '''pgm := line pgm | EOF
line := lin_num stmt
stmt := asgmnt | if | print | goto | stop
asgmnt := id = exp
exp := term exp'
exp' := + term | - term | empty
term := id | const
if := IF cond lin_num
cond := term cond'
cond' := < term | = term
print := PRINT id
goto := GOTO lin_num
stop := STOP'''
```

```python
def transform_patterns(patterns, key_regex):
    patterns = re.sub(key_regex, r'{\1}', patterns)
    patterns = re.sub('EOF', '$', patterns)
    patterns = re.sub('empty', '', patterns)
    return list(map(str.strip, patterns.split('|')))
```

```python
rules = dict()
for line in (line.strip() for line in grammar.split('\n')):
    key, patterns = map(str.strip, line.split(':='))
    if key in rules:
        raise Exception('redundant keys')

    rules[key] = patterns
    
key_regex = '|'.join(sorted(rules.keys(), key=lambda t: len(t), reverse=True)).join('()')
rules = {key: transform_patterns(patterns, key_regex) for key, patterns in rules.items()}
rules
```

# First Set

```python
def is_terminate(token):
    return token == '' or token[0] != '{'
def get_non_terminate_name(token):
    return re.findall(r'\{([^\}]+)\}', token)[0]
```

```python
first_set = {key:set() for key in rules.keys()}

len_first_set = -1
len_first_set_new = sum(len(a) for a in first_set.values())

while len_first_set != len_first_set_new:
    len_first_set = len_first_set_new

    for key, patterns in rules.items():
        for pattern in patterns:
            tokens = pattern.split(' ')
            
            ftoken = tokens[0]
            if not tokens[0]:
                first_set[key].add(ftoken)
            elif is_terminate(ftoken):
                first_set[key].add(ftoken)
            else:
                t = get_non_terminate_name(ftoken)
                first_set[key].update(first_set[t])

    len_first_set_new = sum(len(a) for a in first_set.values())
```

```python
for left, right in rules.items():
    print('{left:6}:= {right:40}=> {first_set}'.format(
        left=left, right='| '.join(right), first_set=first_set[left]
    ))
```

# Follow Set

```python
def get_first_set(token):
    if is_terminate(token):
        return {token}
    token = get_non_terminate_name(token)
    return first_set[token]
```

```python
follow_set = {key:set() for key in rules.keys()}
follow_set['pgm'].add('$')

len_follow_set = -1
len_follow_set_new = sum(len(a) for a in follow_set.values())

while len_follow_set != len_follow_set_new:
    len_follow_set = len_follow_set_new
    
    for key, patterns in rules.items():
        for pattern in patterns:
            tokens = pattern.split()
            tokens.append('')
            
            i = len(tokens)-1
            while i >= 0:
                if not is_terminate(tokens[i]):
                    t = get_non_terminate_name(tokens[i])
                    if i < len(tokens)-1 and '' not in get_first_set(tokens[i+1]):
                        break
                    follow_set[t].update(follow_set[key] - {''})
                i -= 1
            
            tokens = tokens[:-1]
            for r in range(len(tokens)-1, 0, -1):
                if not is_terminate(tokens[r-1]):
                    t = get_non_terminate_name(tokens[r-1])
                    follow_set[t].update(get_first_set(tokens[r]) - {''})
    
    len_follow_set_new = sum(len(a) for a in follow_set.values())
```

```python
for left, right in rules.items():
    print('{left:6}:= {right:40}=> {follow_set}'.format(
        left=left, right='| '.join(right), follow_set=follow_set[left]
    ))
```

```python
terminal_set = set()
nonterminal_set = set()

for key, patterns in rules.items():
    for pattern in patterns:
        tokens = pattern.split(' ')
        ftoken = tokens[0]
        
        if is_terminate(ftoken):
            terminal_set.add(ftoken)
    nonterminal_set.add(key)
            
terminal_order = list(reversed(sorted(terminal_set)))
nonterminal_order = list(reversed(sorted(nonterminal_set)))
```
